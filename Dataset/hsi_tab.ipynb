```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
from itertools import product
import os
import colorsys
```

```python
location = 'Report/'
```

```python
deg_range = 15
color_range = 0.05
```

```python
def convert_rgb_hsi(img):
    temp = img.copy()
    # print(temp.shape[0], temp.shape[1], temp.shape[2])
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            # print('{} {}'.format(i, j))
            h, s, v = colorsys.rgb_to_hsv(img[i, j, 0], img[i, j, 1], img[i, j, 2])
            h = h*360
            # print('h: {:.4f} | s: {:.4f} | v: {:.4f}'.format(h, s, v))
            temp[i, j, 0] = h
            temp[i, j, 1] = s
            temp[i, j, 2] = v
    return temp
```

```python
def count_px(img, deg_range, color_range):
    
    hsv_count = np.array([[[0 for i in range(0, int(1/color_range), 1)] for i in range(0, int(1/color_range), 1)] for i in range(0, 360, deg_range)])
    
    for i in range(0, img.shape[0]):
        for j in range(0, img.shape[1]):
            if(sum(img[i, j, :]!=0)):
                h_val = img[i, j, 0]
                s_val = img[i, j, 1]
                v_val = img[i, j, 2]
                if(h_val==360):
                    h_val = 0
                h_idx = int(h_val/deg_range)
                if(s_val==1):
                    s_idx = int(s_val/color_range)-1
                else:
                    s_idx = int(s_val/color_range)
                if(v_val==1):
                    v_idx = int(v_val/color_range)-1
                else:
                    v_idx = int(v_val/color_range)

                # print('h: {}, s: {}, v: {}'.format(h_idx, s_idx, v_idx))
                hsv_count[h_idx, s_idx, v_idx] = hsv_count[h_idx, s_idx, v_idx]+1
    
    count_list = []
    for i in range(0, int(360/deg_range)):
        for j in range(0, int(1/color_range)):
            for k in range(0, int(1/color_range)):
                count_list.append(hsv_count[i, j, k])
    
    return count_list
```

```python
def transform_tab(tab, tab_init, top_n, filename, label):
    top = tab.sort_values('proportion', ascending=False).iloc[0:7].drop('count', axis=1)
    temp = tab_init.copy()
    temp['filename'] = [filename]
    for i in range(1, top_n+1):
        # print(i)
        temp['h_range_{}'.format(i)] = [top.iloc[i-1, 0]]
        temp['s_range_{}'.format(i)] = [top.iloc[i-1, 1]]
        temp['v_range_{}'.format(i)] = [top.iloc[i-1, 2]]
        temp['proportion_{}'.format(i)] = [top.iloc[i-1, 3]]
    temp['label'] = label
    return temp
```

```python
top7_tab_init = pd.read_csv('Report/top7_hsv_count_tab.csv')
top7_tab = top7_tab_init.copy()
count_tab_init = pd.read_csv('Report/count_hsv_tab_init.csv')
```

```python
location = 'C:\\Users\\Asus\\Documents\\Thesis\\Me\\Dataset\\'
location = ''
location = os.path.join(location, 'european-street-fashion-output/')
locationList = ['european-street-fashion-output/', 'harajuku-kawaii-street-fashion-output/', 'thai-street-fashion-output/']
# locationList = ['thai-street-fashion-output/']
for location in locationList:
    for file_name in os.listdir(location):
        print(location+file_name)
```

```python
for location in locationList:
    if('harajuku' in location):
        label = 1
    else:
        label = 0
    for file_name in os.listdir(location):
        if('.jpg' in file_name):
            print(file_name)
            file_loc = location + file_name
            test = cv2.imread(file_loc)
            test = (test)/255
            temp = convert_rgb_hsi(test)
            count_tab = count_tab_init.copy()
            print(file_loc)
            count_list = count_px(temp, deg_range, color_range)
            count_tab['count'] = count_list
            count_tab['proportion'] = count_tab['count']/sum(count_tab['count'])
            temp_top7_tab = transform_tab(count_tab, top7_tab_init, 7, file_loc, label)
            top7_tab = pd.concat([top7_tab, temp_top7_tab], axis=0)
```
